# Orbital CSV önizleme (ilk 20 satır)

`deb_train`, `deb_test` ve `sat` birleşik CSV dosyalarını **pandas** ile okuyup Jupyter’da tablo olarak gösterir.

**Not:** Bu notebook’u `notebooks/` içinden veya proje kökünden açabilirsiniz; kod kökteki veya `data/processed/` altındaki CSV’leri otomatik bulur.

```python
from pathlib import Path

import pandas as pd
from IPython.display import display, Markdown


def project_root() -> Path:
    cwd = Path.cwd().resolve()
    if (cwd / "deb_train_combined.csv").exists():
        return cwd
    if (cwd.parent / "deb_train_combined.csv").exists():
        return cwd.parent
    if (cwd / "data" / "processed" / "deb_train_combined.csv").exists():
        return cwd
    if (cwd.parent / "data" / "processed" / "deb_train_combined.csv").exists():
        return cwd.parent
    return cwd


def load_combined(name: str) -> pd.DataFrame:
    root = project_root()
    for folder in [root, root / "data" / "processed"]:
        path = folder / name
        if path.is_file():
            return pd.read_csv(path)
    raise FileNotFoundError(f"Bulunamadı: {name} (aranan kök: {root})")


ROOT = project_root()
print("Proje kökü:", ROOT)
```

```python
N = 20

df_train = load_combined("deb_train_combined.csv")
df_test = load_combined("deb_test_combined.csv")
df_sat = load_combined("sat_combined.csv")

for baslik, df in [
    ("deb_train (ilk 20)", df_train),
    ("deb_test (ilk 20)", df_test),
    ("sat (ilk 20)", df_sat),
]:
    display(Markdown(f"### {baslik}"))
    display(df.head(N))
```
